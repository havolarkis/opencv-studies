## Loading image file

Load an color image and converting BGR=>RGB

Using matplotlib to plot

```python
from matplotlib import pyplot as plt
import numpy as np
# import cv2 # already imported above

img = cv2.imread('birds.jpg')
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(imgRGB)
plt.title('birds')
plt.show()
```

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imgGray3Channels = cv2.cvtColor(imgGray, cv2.COLOR_GRAY2RGB)

plt.imshow(imgGray3Channels)
plt.title('gray-birds')
plt.show()
```

```python

imgBinding = np.hstack((imgRGB, imgGray3Channels))

plt.imshow(imgBinding)
plt.title('binding-birds')
plt.show()
```
