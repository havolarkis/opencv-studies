# Testing face detection

using multiple cascade models

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

print("opencv-",cv2.__version__)

faceImg = cv2.imread('people.jpg')
plt.imshow(cv2.cvtColor(faceImg, cv2.COLOR_BGR2RGB))
plt.title('people')
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_eye.xml')

faceGray = cv2.cvtColor(faceImg, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(faceGray)
for (x,y,w,h) in faces:
    cv2.rectangle(faceImg,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = faceGray[y:y+h, x:x+w]
    roi_color = faceImg[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        
faceMarked = cv2.cvtColor(faceImg, cv2.COLOR_BGR2RGB)
cv2.imwrite('marked_people.jpg',faceMarked)
        
plt.imshow(faceMarked)
plt.title('people')
plt.show()
```
